```python
fig1 = make_subplots(rows=1, cols=2, specs=[[{}, {}]], shared_xaxes=True,
                    shared_yaxes=False, vertical_spacing=0.1)
```

```python
fig1.append_trace(go.Scatter(
    x= y_ev_bottom , y= x1,
    mode='lines+markers',
    #line_color='rgb(128, 0, 128)',
    name='EV Percentage',
), 1, 2)
```

```python
def add_bars(figure, x_values, y_values, type_name ):
  figure.add_trace(go.Bar(
    x= x_values,  # Replace other_values with your distribution values
    y=y_values,    # Replace categories with your category labels
    marker=dict(
        #color='rgba(115, 171, 255, 0.6)',  # Adjust color for distribution bars
        #line=dict(color='rgba(115, 171, 255, 1.0)', width=2),
    ),
    name= type_name,  # Name for distribution bars
    orientation='h',
));
```

```python
add_bars(fig1, df_bottom['Diesel'], x1, 'Disel')
add_bars(fig1, df_bottom['Electric'], x1, 'Electric')
add_bars(fig1, df_bottom['Flex_Fuel'], x1, 'Flex_Fuel')
add_bars(fig1, df_bottom['Gasoline'], x1, 'Gasoline')
add_bars(fig1, df_bottom['Gasoline_Hybrid'], x1, 'Gasoline Hybrid')
add_bars(fig1, df_bottom['Hydrogen'], x1, 'Hydrogen')
add_bars(fig1, df_bottom['Natural_Gas'], x1, 'Natural Gas')
add_bars(fig1, df_bottom['PHEV'], x1, 'PHEV')
```

```python
fig1.update_layout(
    title='Fuel Type distribution for bottom 20',
    yaxis=dict(
        showgrid=False,
        showline=False,
        showticklabels=True,
        domain=[0, 0.85],
    ),
    yaxis2=dict(
        showgrid=False,
        showline=True,
        showticklabels=False,
        linecolor='rgba(102, 102, 102, 0.8)',
        linewidth=2,
        domain=[0, 0.85],
    ),
    xaxis=dict(
        zeroline=False,
        showline=False,
        showticklabels=True,
        showgrid=True,
        domain=[0, 0.42],
    ),
    xaxis2=dict(
        zeroline=True,
        showline=True,
        showticklabels=True,
        showgrid=True,
        domain=[0.47, 1],
        side='top',
        dtick=25000,
        #autorange='reversed'
    ),
    legend=dict(x=10, y=1.038, font_size=10),
    margin=dict(l=100, r=20, t=70, b=70),
    paper_bgcolor='rgb(248, 248, 255)',
    plot_bgcolor='rgb(248, 248, 255)',
    barmode='relative'
);

#fig1.update_layout( yaxis2={'categoryorder':'total descending'})
```
